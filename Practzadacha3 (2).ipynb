Задание 3, вариант 1 из Демченко

$$\frac{du}{dt} - 2\frac{du}{dx} = 0, 0<t\leq1, 0\leq x<1$$
$$u(x,0) = cos x$$
$$u(1,t) = cos(1+2t)$$

Разностная схема:

$$D_n = {(x_l,t^n): x_l=hl, hL = 1, l = 0,...L, t^n = n\tau, N\tau = 1, n = 0,...N}$$
$$u^{n+1}_l = u_l^n + 2\frac{\tau}{h}(u^n_{l+1}-u^n_l), l = 0,...L-1, n = 0, ...N-1$$
$$u^0_l=cos x_l, l = 0,...L$$
$$u^n_L = cos(1+2t^n), n = 1,...N$$

Шаблон - правый уголок

Условие устойчивости: $\frac{2\tau}{h}\leq1$

Исследовав схему на аппроксимацию нашли, что она первого порядка по $\tau, h$

```python
#Удовлетворяют условию устойчивости
h = 0.01
L = 1/h
tau = 0.002
N = 1/tau
```

```python
x_l = [h*l for l in range(0,int(L+1))]
tn = [tau*n for n in range(0, int(N+1))]
```

```python
un_l = [[0]*int(L+1) for n in range(0,int(N+1))]
```

```python
from math import cos
```

```python
for l in range(0,int(L+1)):
    un_l[0][l] = cos(x_l[l])
```

```python
for n in range(1,int(N+1)):
    un_l[n][int(L)] = cos(1+2*tn[n])
```

```python
for t in range(1,int(N+1)):
    for l in range(0,int(L)):
        un_l[t][l] = un_l[t-1][l]+2.*tau/h*(un_l[t-1][l+1]-un_l[t-1][l])
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import matplotlib.animation as animation
%matplotlib notebook
```

```python

plt.style.use('dark_background')
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

# animate the time data
k = 0
def animate(i):
    global k
    x = un_l[k]
    k += 1
    ax1.clear()
    plt.plot(x_l,x,color='cyan')
    plt.grid(True)
    plt.ylim([-1,1])
    plt.xlim([0,1])

anim = animation.FuncAnimation(fig,animate,frames=360,interval=20)
plt.show()
# по оси х  у нас расстояние, по у - значение функции
```

```python
ul_n = [[0]*int(N+1) for n in range(0,int(L+1))]
for l in range(0,int(L+1)):
    ul_n[l][0] = cos(x_l[l])
for n in range(1,int(N+1)):
    ul_n[int(L)][n] = cos(1+2*tn[n])
for t in range(1,int(N+1)):
    for l in range(0,int(L)):
        ul_n[l][t] = ul_n[l][t-1]+2.*tau/h*(ul_n[l+1][t-1]-ul_n[l][t-1])
```
